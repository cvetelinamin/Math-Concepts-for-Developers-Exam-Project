# Credit Risk Scoring System

## 1.What is credit risk and why is it important to create an accurate model?

Credit risk refers to the possibility that a borrower will fail to repay a loan or meet contractual obligations, causing financial losses or gains to the lender. It is crucial for lenders and financial institutions to assess this risk accurately to make informed lending decisions.

An accurate credit risk model helps:

Minimize losses by identifying high-risk borrowers.
Optimize loan approval processes, ensuring that credit is extended to those who are likely to repay.
Improve profitability by balancing the risk and reward of lending decisions.

When assessing customer credit risk, it is crucial to consider several vital features that can impact the creditworthiness of borrowers in credit risk prediction.

1. **Basic Features**<br>
a. **Loan amount:** The loan is a critical factor in determining credit risk. Generally, the more significant the loan amount, the higher the risk.<br>
b. **Loan term:** The length of the loan term is another vital consideration. Longer loan terms may be riskier because the borrower is more likely to default over the extended repayment period.<br>
c. **Interest rate:** The interest rate is a critical factor that impacts the credit risk of the loan. Higher interest rates may make it more challenging for the borrower to repay the loan, and they may be more likely to default.<br>
d. **Payment history:** The borrower's payment history is crucial in determining credit risk. If the loan applicant has a history of late or missed payments, it suggests that they may have difficulty repaying the loan.<br>
2. **Customer default history**<br>
a. **The number of past defaults:** The number of times a customer has defaulted on a loan or credit product is an essential indicator of credit risk.<br>
b. **Time since the last default:** The time since the customer's previous default can also be an essential indicator of credit risk.<br>
3. **Customer demography**<br>
a. **Age:** Age is a significant factor in assessing credit risk, as younger customers may need more experience managing finances and have a limited credit history.<br> 
b. **Income:** Income is crucial in determining creditworthiness, as higher-income customers are typically seen as less risky borrowers.<br>
c. **Employment status:** The employment status of a customer can also impact credit risk. Customers with stable, full-time employment are typically viewed as less risky borrowers than those with part-time or self-employment income.
4. **City level features**<br>
a. **Economic conditions:** Economic conditions in a particular city can significantly impact credit risk. Factors such as unemployment rates, inflation rates, and GDP growth can affect the ability of borrowers to repay their loans.<br>
b. **Demographics:** Demographic factors such as Age, education level, and income can impact credit risk. For example, younger borrowers or those with lower levels of education may be considered at higher risk due to their lack of financial experience.
